## Working with Missing Data in Pandas

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

### Filling missing values using fillna(), replace() and interpolate()

```python
data = {'name': ['steve','john','richard','sarah','randy'
                 ,'micheal','julie']
        ,'age': [20,21,33,23,42,38,22]         
        ,'gender':['male','male','male','female','male'
                 ,'male','female']
        ,'rank':[2,1,3,5,4,7,6]         
                 }
```

```python
ranking_df = DataFrame(data)
```

```python
ranking_df
```

```python
ranking_df.iloc[2:5,1] = np.nan
```

```python
ranking_df.iloc[3:6,3] = np.nan
```

```python
ranking_df
```

```python
ranking_df.isnull().sum()
```

```python
ranking_df.notnull().sum()
```

```python
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
ranking_df
```

```python
test = ranking_df.fillna(0)
```

```python
test = ranking_df.bfill()
test
```

```python
test = ranking_df.ffill()
test
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how='all')
```
